```python
from datetime import date
```

```python
page = requests.get("https://cookpolitical.com/ratings/senate-race-ratings")
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
# write soup to html to save when future ratings changes would overwrite
with open("senate-"+str(date.today())+".html", "w") as file:
    file.write(str(soup))
```

```python
soup.find_all('ul', class_='ratings-detail-page-table-7-column-ul')
```

```python
categories = soup.find_all('ul', class_='ratings-detail-page-table-7-column-ul')
```

```python
dem=['DSOLID_DEMOCRAT',
'DLIKELY_DEMOCRAT',
'DLEAN_DEMOCRAT',
'DTOSS_UP',
'DLEAN_REPUBLICAN',
'DLIKELY_REPUBLICAN',
'DSOLID_REPUBLICAN','RSOLID_DEMOCRAT',
'RLIKELY_DEMOCRAT',
'RLEAN_DEMOCRAT',
'RTOSS_UP',
'RLEAN_REPUBLICAN',
'RLIKELY_REPUBLICAN',
'RSOLID_REPUBLICAN']
```

```python
catLists=[]
```

```python
for cat in categories:
    catLists.append([elem.get_text().strip('\n') for elem in cat.select("li")])
```

```python
catLists
```

```python
import pandas as pd
```

```python
df= pd.DataFrame(catLists).transpose()
```

```python
df.columns=dem
```

```python
df
```

```python
s = pd.melt(df).dropna()
```

```python
s['state']=s['value']
```

```python
s['state']=s['state'].str.slice(stop=2)
```

```python
s['name']=s['value']
```

```python
s['name']=s['name'].str.slice(start=3)
```

```python
s['held']=s['variable']
```

```python
s['held']=s['held'].str.slice(stop=1)
```

```python
s['projection']=s['variable']
```

```python
s['projection']=s['projection'].str.slice(start=1)
```

```python
s=s.drop(['variable','value'],axis=1)
```

```python
s['state_unique']=s['state']+"20"
```

```python
s.iloc[14].state=s.iloc[14].state="GAsp"
```

```python
s.iloc[14].state_unique=s.iloc[14].state_unique="GAsp"
```

```python
noelections=pd.read_csv('senate-noelections.csv')
```

```python
senate=pd.concat([s,noelections],axis=0)
```

```python
pd.set_option('display.max_rows', 500)
```

```python
projOrder = ['noelectiondem','SOLID_DEMOCRAT','LIKELY_DEMOCRAT','LEAN_DEMOCRAT','TOSS_UP','LEAN_REPUBLICAN','LIKELY_REPUBLICAN','SOLID_REPUBLICAN','noelectionrep']
```

```python
senate['projection']=pd.Categorical(senate.projection, 
                      categories=projOrder)
```

```python
senate.sort_values('projection', inplace=True)
```

```python
senatedata = pd.read_csv("senateall.csv")
```

```python
senatedata.head()
```

```python
senate = senate.merge(senatedata, on="state_unique")
```

```python
senateDict = {
    "noelectiondem":"No election",
    "noelectionrep": "No election",
    "SOLID_DEMOCRAT":"Solid Democrat",
   "LIKELY_DEMOCRAT":"Likely Democrat",
    "LEAN_DEMOCRAT":"Lean Democrat",
    "TOSS_UP":"Toss Up",
    "LEAN_REPUBLICAN":"Lean Repbulican",
    "LIKELY_REPUBLICAN":"Likely Repbulican",
    "SOLID_REPUBLICAN":"Solid Republican"
}
```

```python
senate['rating']=senate['projection']
```

```python
senate=senate.replace({"rating":senateDict})
```

```python
senate=senate.reset_index(drop=True)
```

```python
senate['index']=senate.index
```

```python
senate.to_json('senate.json',orient='records')
```
